```python
#load data
import scipy.io as sio
import numpy as np
import matplotlib.pyplot as plt

data = sio.loadmat('ex6data1.mat')
X = data['X']
y = data['y']
print X[:5]
print y[:5]

label0 = np.where(y == 0)
plt.scatter(X[label0,0],X[label0,1],marker='x',color = 'r',label = '0')
label1 = np.where(y == 1)
plt.scatter(X[label1,0],X[label1,1],marker='o',color = 'b',label = '1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend(loc = 'upper left')
plt.show()


```

```python
#Training Linear SVM on dataset1
from sklearn import svm
#linearSVC
C = 1.0 # SVM regularization parameter
clf = svm.LinearSVC(C=C) #model
clf.fit(X,y.ravel()) #train
#prediction = clf.predict([[1.5,2.0],[1.8,3.0],[4.0,4.2]])
#print prediction



```

```python
# 咱们先定义一个一个函数来画决策边界  
def plot_decision_boundary(pred_func,X,y):  
  
    # 设定最大最小值，附加一点点边缘填充  
    x_min, x_max = X[:, 0].min() - 0.1, X[:, 0].max() + 0.1  
    y_min, y_max = X[:, 1].min() - 0.1, X[:, 1].max() + 0.1  
    h = 0.01  
    
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))  
  
    # 用预测函数预测一下  
    Z = pred_func(np.c_[xx.ravel(), yy.ravel()])  
    Z = Z.reshape(xx.shape)  
    #plt.xlim(x_min,x_max)
    #plt.ylim(y_min,y_max)
    # 然后画出图  
    plt.contour(xx, yy, Z, colors='b', linewidth=.5)  
    plt.scatter(X[:, 0], X[:, 1], c=y) 
    
plot_decision_boundary(lambda x:clf.predict(x),X,y)
plt.title("LinearSVM")  
plt.show()
    
```

```python
#load data2
data2 = sio.loadmat('ex6data2.mat')
X2 = data2['X']
y2 = data2['y']
print X2[:5]
print y2[:5]

label0 = np.where(y2 == 0)
plt.scatter(X2[label0,0],X2[label0,1],marker='x',color = 'r',label = '0')
label1 = np.where(y2 == 1)
plt.scatter(X2[label1,0],X2[label1,1],marker='o',color = 'b',label = '1')
plt.xlabel('X1')
plt.ylabel('X2')
plt.legend(loc = 'upper left')
plt.show()
```

```python
#Training SVM with RBF Kernel (Dataset 2)
from sklearn import svm
clf2 = svm.SVC(kernel='rbf',gamma=1.0,C=1.0)
clf2.fit(X2,y2.ravel())


```

```python
plot_decision_boundary(lambda x:clf2.predict(x),X2,y2)
plt.title("RBF-SVM")  
plt.show()
```
